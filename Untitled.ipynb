```python
import streamlit as st
import yfinance as yf
import google.generativeai as genai
import pandas as pd

# --- Configuration ---
st.set_page_config(page_title="GOOGL Deep Dive", page_icon="📊", layout="wide")

# --- Sidebar: API Key Input ---
st.sidebar.header("🔑 App Configuration")
api_key = st.sidebar.text_input("Enter your Gemini API Key:", type="password")
st.sidebar.markdown("""
*Don't have a key? Get one for free at [Google AI Studio](https://aistudio.google.com/).*
""")

# --- Main Dashboard ---
st.title("📊 Alphabet Inc. (GOOGL) AI Analysis")
st.markdown("A complete technical, fundamental, and sentiment breakdown powered by AI.")

# Target Ticker
TICKER = "GOOGL"

@st.cache_data(ttl=3600) # Cache data for 1 hour to speed up reloads
def get_stock_data(ticker_symbol):
    stock = yf.Ticker(ticker_symbol)
    hist = stock.history(period="1y")
    info = stock.info
    news = stock.news
    return stock, hist, info, news

with st.spinner("Fetching market data..."):
    stock, hist, info, news = get_stock_data(TICKER)

# --- 1. Top Level Metrics ---
col1, col2, col3, col4 = st.columns(4)
current_price = hist['Close'].iloc[-1]
prev_price = hist['Close'].iloc[-2]
price_change = current_price - prev_price
pct_change = (price_change / prev_price) * 100

col1.metric("Current Price", f"${current_price:.2f}", f"{price_change:.2f} ({pct_change:.2f}%)")
col2.metric("Market Cap", f"${info.get('marketCap', 0) / 1e12:.2f} Trillion")
col3.metric("P/E Ratio", f"{info.get('trailingPE', 'N/A')}")
col4.metric("52 Week High", f"${info.get('fiftyTwoWeekHigh', 'N/A')}")

# --- 2. Technical Chart ---
st.subheader("📈 1-Year Price Action & Moving Averages")
# Calculate simple moving averages (SMA)
hist['50_Day_SMA'] = hist['Close'].rolling(window=50).mean()
hist['200_Day_SMA'] = hist['Close'].rolling(window=200).mean()

# Plotting using Streamlit's native chart for speed
chart_data = hist[['Close', '50_Day_SMA', '200_Day_SMA']]
st.line_chart(chart_data)

# --- 3. AI Deep Dive ---
st.divider()
st.subheader("🧠 AI Analyst Report")

if not api_key:
    st.warning("👈 Please enter your Gemini API Key in the sidebar to generate the AI analysis.")
else:
    if st.button("Generate Comprehensive AI Report"):
        with st.spinner("AI is analyzing charts, fundamentals, and breaking news..."):
            try:
                # Configure AI
                genai.configure(api_key=api_key)
                model = genai.GenerativeModel('gemini-2.5-flash') # Using the latest fast model

                # Prepare the data payload for the AI
                recent_news = "\n".join([f"- {n.get('title')}" for n in news[:5]])
                
                prompt = f"""
                You are a Senior Equity Research Analyst. Provide a comprehensive analysis of Alphabet Inc. (GOOGL).
                
                Here is the current data I have fetched for you:
                - Current Price: ${current_price:.2f}
                - 50-Day SMA: ${hist['50_Day_SMA'].iloc[-1]:.2f}
                - 200-Day SMA: ${hist['200_Day_SMA'].iloc[-1]:.2f}
                - P/E Ratio: {info.get('trailingPE', 'N/A')}
                
                Recent News Headlines:
                {recent_news}
                
                Please provide a structured report with the following sections:
                1. **Technical Outlook**: Analyze the price against the 50-day and 200-day moving averages.
                2. **Fundamental Quick Take**: Briefly comment on the valuation based on the P/E ratio.
                3. **News Sentiment**: What is the overall vibe of the recent news?
                4. **Final Verdict**: Give a clear Bullish, Bearish, or Neutral stance with a one-sentence justification.
                """
                
                response = model.generate_content(prompt)
                st.markdown(response.text)
                
            except Exception as e:
                st.error(f"API Error: {e}. Please check your API key and try again.")
```
